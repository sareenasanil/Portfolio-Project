```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('movies.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
for col in data.columns:
    missing=np.mean(data[col].isnull())
    print('{} - {}%'.format(col,missing))
```

```python
# data types of cols
data.dtypes
```

```python
#data type change
data['budget']=data['budget'].astype('int64')
data['gross']=data['gross'].astype('int64')
```

```python
data.head()
```

```python
data['year']=data['released'].apply(lambda x : x.split('-')[0])
# df['yearcorrect']=data['released'].astype(str).str[:4]
```

```python
data.head()
```

```python
data.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#drop any duplicates
data['company'].drop_duplicates().sort_values(ascending=False)
#data.drop_duplicates()
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
plt.scatter(x=data['budget'],y=data['gross'])
```

```python
datanum=data
for cols in datanum.columns:
    if (datanum[cols].dtype=='object'):
        datanum[cols]=datanum[cols].astype('category')
        datanum[cols]=datanum[cols].cat.codes
datanum.head()
```

```python
data.head()
```

```python
fig= plt.figure(figsize=(10,5))
ax=fig.add_axes([0,0,1,1])
sns.heatmap(datanum.corr(),annot=True,ax=ax)
```

```python
corrmat=datanum.corr()
corr_pairs=corrmat.unstack()
corr_pairs.head()
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs.head()
```

```python
#high correlation
sorted_pairs[sorted_pairs>0.5].head()
```
